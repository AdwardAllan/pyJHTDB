First, construct the points where we're gonna compute the various fields.
The only nonisotropic grid is the one for the channel flow, therefore I'm generating the x and z coordinates randomly, but keeping y in check somewhat.
This shouldn't introduce any bias for the isotropic datasets.

```python
import numpy as np
import pyJHTDB
from pyJHTDB.dbinfo import channel as info

npoints = 2**10
nskip = 2

x = np.random.random(
        size = (npoints, int(0.55*info['ny']/nskip)-1, 3)).astype(np.float32)
x[..., 0] *= info['lx']
x[..., 1] *= info['dy'][::nskip][None, :x.shape[1]]
x[..., 1] += info['ynodes'][::nskip][None, :x.shape[1]]
x[..., 2] *= info['lz']

time = np.random.choice(info['time'])
```

Now I'm actually calling the database, and requesting the velocity gradient.
Afterwards, I'm averaging over the first axis of x, since the y dependency is in the second axis.
For now, only M1Q4, M2Q8 and M2Q14 are implemented for the channel dataset, therefore they're the only ones I'm using

```python
from pyJHTDB import libJHTDB

interp_info = [(208, 'M2Q8'),
               (214, 'M2Q14')]
divu = []
gradu = []
uval = []

lJHTDB = libJHTDB()
lJHTDB.initialize()
for ii in interp_info:
    uval.append(lJHTDB.getData(
                0.0,
                x,
                sinterp = ii[0],
                tinterp = 0,
                data_set = info['name'],
                getFunction = 'getVelocity'))
    current_gradu = lJHTDB.getData(
                0.0,
                x,
                sinterp = ii[0],
                tinterp = 0,
                data_set = info['name'],
                getFunction = 'getVelocityGradient')
    gradu.append(current_gradu)
    divu.append(np.average(np.abs(
                        current_gradu[..., 0] +
                        current_gradu[..., 4] +
                        current_gradu[..., 8]), axis = 0) / 
                np.average(np.sqrt(
                        current_gradu[..., 0]**2 +
                        current_gradu[..., 4]**2 +
                        current_gradu[..., 8]**2), axis = 0))
lJHTDB.finalize()
```

This is the divergence error.

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111)
for i in range(len(divu)):
    ax.plot(info['ynodes'][::nskip][:divu[i].shape[0]],
            divu[i],
            label = interp_info[i][1])
ax.legend(loc = 'best')
ax.set_yscale('log')
ax.set_xlabel('$y$')
ax.set_ylabel('relative divergence error')
```

Now the regular error, measured by difference between the 2 successive methods.

```python
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111)
def plot_component_error(
        err,
        size,
        component,
        label,
        dashes = (None, None)):
    ax.plot(info['ynodes'][::nskip][:err.shape[1]],
            (np.average(err[..., component], axis = 0) /
             np.average(size[..., component], axis = 0)),
            label = label,
            dashes = dashes,
            linewidth = 2)

uerr  = np.abs(uval[0] - uval[1])
usize = np.abs(uval[1])
plot_component_error(uerr, usize, 0, '$u_x$')
plot_component_error(uerr, usize, 1, '$u_x$')
plot_component_error(uerr, usize, 2, '$u_x$')
duerr  = np.abs(gradu[0] - gradu[1])
dusize = np.abs(gradu[1])
plot_component_error(duerr, dusize, 0, '$\\partial_x u_x$', dashes = (4,4))
plot_component_error(duerr, dusize, 1, '$\\partial_y u_x$', dashes = (2,2))
plot_component_error(duerr, dusize, 2, '$\\partial_z u_x$', dashes = (5,5))
plot_component_error(duerr, dusize, 3, '$\\partial_x u_y$', dashes = (4,4))
plot_component_error(duerr, dusize, 4, '$\\partial_y u_y$', dashes = (2,2))
plot_component_error(duerr, dusize, 5, '$\\partial_z u_y$', dashes = (5,5))
plot_component_error(duerr, dusize, 6, '$\\partial_x u_z$', dashes = (4,4))
plot_component_error(duerr, dusize, 7, '$\\partial_y u_z$', dashes = (2,2))
plot_component_error(duerr, dusize, 8, '$\\partial_z u_z$', dashes = (5,5))
ax.legend(loc = 'best')
ax.set_yscale('log')
ax.set_xlabel('$y$')
ax.set_ylabel('relative interpolation error')
ax.legend(loc = 'best')
```
